#Feature Relationship

```python
df.columns
```

```python
sns.scatterplot(data=df,
                x='Speed_mph',
                y='Gforce',
                hue='Year_Introduced')
plt.title('Scatterplot')
plt.xlabel('Speed mph')
plt.ylabel('Gforce')
plt.show()
```

```python
correlation_matrix = df.corr()

# Plot correlation heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='Type_Main')
plt.show()
```

#Ask Question about the Data

- What are the Locations with the fastest roller coasters ?

```python
df['Location'].value_counts()
```

```python
ax=df.query('Location !="Other"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count >=10')\
.sort_values('mean')\
.plot(kind='barh', figsize=(12,5), title='Average Coaster Speed by Locaiton')
ax.set_xlabel('Averge Speed by Location')
plt.show()
```
